```python
import os
import subprocess
import tkinter as tk
from nbconvert import PythonExporter
import nbformat

SCRIPT_DIR = os.getcwd()

def convert_and_run_notebook(notebook_name):
    """Convert Jupyter Notebook to Python script and run it."""
    notebook_path = os.path.join(SCRIPT_DIR, notebook_name)
    
    if os.path.exists(notebook_path):
        try:
            # Load the .ipynb file
            with open(notebook_path, 'r', encoding='utf-8') as f:
                nb = nbformat.read(f, as_version=4)
            
            # Convert notebook to Python script
            exporter = PythonExporter()
            python_script, _ = exporter.from_notebook_node(nb)
            
            # Save converted script
            py_file = notebook_name.replace(".ipynb", ".py")
            py_path = os.path.join(SCRIPT_DIR, py_file)
            with open(py_path, 'w', encoding='utf-8') as f:
                f.write(python_script)
            
            # Run the converted script
            subprocess.Popen(["python", py_path], creationflags=subprocess.CREATE_NEW_CONSOLE)

        except Exception as e:
            tk.messagebox.showerror("Error", f"Failed to convert or execute {notebook_name}\n{e}")
    else:
        tk.messagebox.showerror("Error", f"File {notebook_name} not found!")

# Function to close the application
def close_application():
    root.destroy()

# GUI Setup
root = tk.Tk()
root.title("DDoS Detection System")
root.geometry("600x350")
root.configure(bg="#f0f0f0")

# Title Label
tk.Label(root, text="DDoS Detection System", font=("Times New Roman", 16, "bold"), bg="#f0f0f0").pack(pady=20)

# Frame for buttons
frame = tk.Frame(root, padx=30, pady=20, relief=tk.RIDGE, bd=2, bg="#ffffff")
frame.pack(pady=15)

# Buttons to convert and execute notebooks
tk.Button(frame, text="Model Trainer and Evaluator", 
          command=lambda: convert_and_run_notebook("evaluator.ipynb"), 
          width=30, font=("Times New Roman", 12)).pack(pady=10)

tk.Button(frame, text="Performance Analyzer", 
          command=lambda: convert_and_run_notebook("performance_analyzer.ipynb"), 
          width=30, font=("Times New Roman", 12)).pack(pady=10)

# Close button outside the frame
tk.Button(root, text="Close", command=close_application, width=30, font=("Times New Roman", 12)).pack(pady=10)

root.mainloop()
```
